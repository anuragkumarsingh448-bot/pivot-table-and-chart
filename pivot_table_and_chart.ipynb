Question no. 1 What is purpose of pivot table in excel?

Answer: The primary purpose of a pivot table is to summarize, analyze, explore, and present large amount of data quickly. Instead of writing complex formulas, it allows you to dynamically reorganize rows and columns for adding, average or count



Question no. 4 Explain the difference between sorting and filtering with an example

Answer: Sorting change the order of your data without hiding any rows.

Filtering temporarily hides rows that do not meet specific criteria, showing you only the subset of data you care about.

Example Sorting: you want to see which specific transactions generated the absolute highest revenue

select the sale column and sort it from Largest to Smallest

Filtering: you want to analyze how your oranges are performing across all regions

Click the filter dropdown arrow on the product column header uncheck "select all", check only orange and click ok

Question no. 7 Why are dashboards important for decision-making in business analysis?

Answer: Dashboard are critical because they turn complex, scattered datasets into a centralized visual story.      

Instead of scrolling through thousands of rows of transactions, executives can immediately spot performance trends, High-performing regions, or lagging products.
features like slicers allow decision-makers to ask and answer their own questions on the fly without needing a data analyst to generate a new report.

They save time by consolidating key performance indicators into a single, clean workspace.
